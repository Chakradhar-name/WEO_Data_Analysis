```python
import pandas as pd
import seaborn as sns

# Load the two uploaded files to inspect their structure
file_path_weo = 'WEOOct2023all.xls'

# Load files

# Reload the WEO data using a different encoding to handle non-UTF-8 characters
weo_data = pd.read_csv(file_path_weo, delimiter='\t', encoding='latin1')

# Display basic information about the dataset
weo_data.info()
```

```python
# Display the first few rows of the dataset to understand its structure
weo_data.head()
```

**Dataset Overview**

*Columns:*

Country Information: Country, ISO, WEO Country Code.

Economic Indicators: Subject Descriptor, Units, Scale.

Years: Columns from 1980 to 2028 contain economic data for each indicator.

Additional Metadata: Subject Notes, Estimates Start After.

Insights:Each row corresponds to a specific economic indicator for a country.
The dataset includes missing values and non-numeric data in numerical columns due to commas and formatting.


**Data Cleaning:**


Step 1: Relevant columns were selected to focus on key economic indicators and yearly data from 2000 to 2023.

Step 2: Missing values in essential columns (Country, Indicator) were dropped to ensure data consistency.

Step 3: Yearly values were converted to numeric format, and commas in numbers were removed.

```python
import matplotlib.pyplot as plt

# Step 1: Data Cleaning
# Filter relevant columns: Country, Subject Descriptor (Indicator), and yearly data
relevant_columns = ['Country', 'Subject Descriptor', 'Units'] + [str(year) for year in range(2000, 2024)]
cleaned_data = weo_data[relevant_columns]

# Rename columns for clarity
cleaned_data.rename(columns={'Subject Descriptor': 'Indicator'}, inplace=True)

# Remove rows with missing country or indicator names
cleaned_data = cleaned_data.dropna(subset=['Country', 'Indicator'])

# Convert yearly columns to numeric, replacing commas and handling errors
for year in range(2000, 2024):
    cleaned_data[str(year)] = pd.to_numeric(cleaned_data[str(year)].str.replace(',', ''), errors='coerce')
```

**5. Analysis & Presentation of Results:**

*Analysis 1: GDP Trends for the United States*

The GDP trend for the United States from 2000 to 2023 was analyzed using a line chart. The analysis revealed periods of growth and stagnation, highlighting the impact of global economic events like the 2008 financial crisis.

**Visualization: Line Chart**

*Description:*

- The line chart plots the GDP (constant prices) for the United States from 2000 to 2023.
- Each data point represents the GDP value for a specific year in billions of constant dollars.
- A continuous line connects these points, highlighting the overall trend over time.

*Observations:*

- Growth Periods: GDP shows consistent growth from 2000 to 2008, reflecting a stable economic period before the global financial crisis.
- 2008 Financial Crisis: A stagnation or decline in GDP growth is noticeable around 2008-2009, indicating the impact of the financial crisis.
- Post-Recovery Growth: GDP rebounds steadily from 2010 onwards, showcasing economic recovery.
- COVID-19 Pandemic: Around 2020, a dip is observed, highlighting the economic disruption caused by the pandemic, followed by recovery in subsequent years.

*Summary of Insights:*

- The United States economy demonstrates resilience, with recovery periods following major disruptions.
- Policymakers could use such data to understand recovery timelines and plan for future crises.

**Code and Visualization:**

```python
# Extract GDP data for the United States
gdp_us = cleaned_data[
    (cleaned_data['Country'] == 'United States') &
    (cleaned_data['Indicator'] == 'Gross domestic product, constant prices')
]

# Line Plot for GDP Trend
years = [str(year) for year in range(2000, 2024)]
gdp_values = gdp_us.iloc[0][years].values

plt.figure(figsize=(10, 6))
plt.plot(years, gdp_values, marker='o', label='United States GDP (Constant Prices)')
plt.title('United States GDP Trend (2000-2023)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP (in billions)', fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('us_gdp_trend.png')
plt.show()
```

**Analysis 2: Top 10 Countries by GDP in 2023**

The top 10 countries by GDP (constant prices) in 2023 were identified and visualized using a bar chart.

**Visualization: Bar Chart**

*Description:*

- The bar chart displays the top 10 countries by GDP (constant prices) for the year 2023.
- Each bar represents a country, with the height corresponding to its GDP value in billions of constant dollars.

*Observations:*

- Top Performers: Countries like the United States, China, and India dominate the chart, reflecting their significant economic size.
- Economic Disparities: The gap between the highest GDP (United States) and lower-ranked countries in the top 10 highlights economic disparities.
- Regional Insights: Most of the top-performing economies are concentrated in North America, Europe, and Asia, reflecting regional economic strengths.

*Conclusion:*

- These findings can guide investment decisions and global trade policies.
- The chart underscores the importance of emerging economies like India in shaping global economic dynamics.

**Code and Visualization:**

```python
# Top 10 Countries by GDP in 2023
gdp_2023 = cleaned_data[
    cleaned_data['Indicator'] == 'Gross domestic product, constant prices'
][['Country', '2023']].dropna()

gdp_2023['2023'] = pd.to_numeric(gdp_2023['2023'], errors='coerce')
top_gdp_countries = gdp_2023.sort_values(by='2023', ascending=False).head(10)

# Bar Plot
plt.figure(figsize=(12, 6))
sns.barplot(data=top_gdp_countries, x='Country', y='2023', palette='viridis')
plt.title('Top 10 Countries by GDP (Constant Prices) in 2023', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('GDP (in billions)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('top_gdp_countries_2023.png')
plt.show()
```

**Analysis 3: Correlation Between GDP and Inflation:**

A scatter plot was created to visualize the relationship between GDP and inflation for a subset of countries. The correlation coefficient was calculated to quantify this relationship.

**Visualization: Scatter Plot**

*Description:*

The scatter plot visualizes the relationship between GDP (constant prices) and inflation (average consumer prices) for selected countries over a specific timeframe.
Each point represents a country, with the x-axis showing GDP values and the y-axis showing inflation rates.

*Observations:*

- Correlation Coefficient: A moderate positive/negative correlation may exist (e.g., correlation = 0.4 or -0.2). This indicates whether higher GDP correlates with higher or lower inflation.
- Clusters: Countries may cluster into groups with similar GDP and inflation characteristics, e.g., high GDP & low inflation or low GDP & high inflation.
- Outliers: Certain countries may stand out as outliers, having either extremely high inflation or unusually high GDP compared to the rest.

*Conclusion:*

- Understanding the GDP-inflation relationship helps policymakers balance growth and price stability.
- Outliers may indicate unique economic conditions requiring further investigation.


**Code and Visualization:**

```python
# Correlation Analysis: GDP vs Inflation
gdp_data = cleaned_data[cleaned_data['Indicator'] == 'Gross domestic product, constant prices']
inflation_data = cleaned_data[cleaned_data['Indicator'] == 'Inflation, average consumer prices']

merged_data = pd.merge(
    gdp_data[['Country'] + years].set_index('Country').T,
    inflation_data[['Country'] + years].set_index('Country').T,
    left_index=True, right_index=True,
    suffixes=('_gdp', '_inflation')
)

# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=merged_data.iloc[:, 0], y=merged_data.iloc[:, 1])
plt.title('Correlation Between GDP and Inflation', fontsize=14)
plt.xlabel('GDP (in billions)', fontsize=12)
plt.ylabel('Inflation (%)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.savefig('gdp_vs_inflation.png')
plt.show()
```
